```python
#88. Merge Sorted Array
class Solution(object):
    def merge(self, nums1, m, nums2, n):
        i = m - 1
        j = n - 1
        k = m + n - 1

        while(i >= 0 and j >= 0):
            if(nums1[i] > nums2[j]):
                nums1[k] = nums1[i]
                i= i-1
            else:
                nums1[k] = nums2[j]
                j-=1
            k-=1
        
        while(j >= 0):
            nums1[k] = nums2[j]
            j-=1
            k-=1
```

```python
#26. Remove Duplicates from Sorted Array
class Solution:
    def removeDuplicates(self, nums):
        if not nums:
            return 0

        k = 1
        for i in range(1, len(nums)):
            if nums[i] != nums[i - 1]:
                nums[k] = nums[i]
                k += 1
        return k
```

```python
@#121. Best Time to Buy and Sell Stock

class Solution(object):
    def maxProfit(self, prices):
        Buy = prices[0]
        Max_profit = 0
        profit = 0
        for Sell in prices:
            profit = Sell - Buy
            Max_profit = max(Max_profit,profit)
            if (profit < 0):
                profit = 0
                Buy = Sell
        return Max_profit
```

```python
#238. Product of Array Except Self

class Solution(object):
    def productExceptSelf(self, nums):
        prefix_arr = [1] * len(nums)
        suffix_arr = [1] * len(nums)
        for i in range(1,len(nums)):
            prefix_arr[i] = prefix_arr[i-1] * nums[i-1]
        for i in range(len(nums)-2,-1,-1):
            suffix_arr[i] = suffix_arr[i+1]*nums[i+1]
        for i in range(len(nums)):
            prefix_arr[i] =prefix_arr[i]*suffix_arr[i]
        return prefix_arr
```

```python
# Two Sum II - Input Array Is Sorted

class Solution(object):
    def twoSum(self, numbers, target):
        left = 0
        right = len(numbers) - 1

        while left < right:
            s = numbers[left] + numbers[right]

            if s == target:
                return [left + 1, right + 1]
            elif s < target:
                left += 1
            else:
                right -= 1
```

```python
#  3643. Flip Square Submatrix Vertically
class Solution(object):
    def reverseSubmatrix(self, grid, x, y, k):
        top = x
        bottom = (x-1)+k
        while top < bottom:
            for i in range(y,y+k):
                grid[top][i],grid[bottom][i] = grid[bottom][i],grid[top][i]
            top +=1
            bottom -=1
        return grid
```

```python
#1886. Determine Whether Matrix Can Be Obtained By Rotation
class Solution(object):
    def findRotation(self, mat, target):
        def rotate(matrix):
            n = len(matrix)

            rotated = [[0] * n for _ in range(n)]
            for i in range(n):
                for j in range(n):
                    rotated[j][n - 1 - i] = matrix[i][j]
            
            return rotated

        for _ in range(4):
            if mat == target:
                return True
            mat = rotate(mat)
        
        return False
```

```python
#3546. Equal Sum Grid Partition I
class Solution(object):
    def canPartitionGrid(self, grid):
        n = len(grid)
        m = len(grid[0])
        totalSum = sum(sum(row) for row in grid)
        topSum = 0
        for i in range(n - 1):
            topSum += sum(grid[i])
            if topSum == totalSum - topSum:
                return True

        leftSum = [0] * m
        
        for i in range(n):
            for j in range(m):
                leftSum[j] += grid[i][j]
        
        curr = 0
        for j in range(m - 1):
            curr += leftSum[j]
            if curr == totalSum - curr:
                return True
        
        return False
```

```python
# 15. 3Sum

class Solution(object):
    def threeSum(self, nums):
        nums.sort()
        result = []

        for i in range(len(nums)):
            if i > 0 and nums[i] == nums[i - 1]:
                continue
            left = i + 1
            right = len(nums) - 1
            while left < right:
                total = nums[i] + nums[left] + nums[right]
                if total == 0:
                    result.append([nums[i], nums[left], nums[right]])

                    while left < right and nums[left] == nums[left + 1]:
                        left += 1

                    while left < right and nums[right] == nums[right - 1]:
                        right -= 1

                    left += 1
                    right -= 1

                elif total < 0:
                    left += 1
                else:
                    right -= 1

        return result
```

```python
# 2946. Matrix Similarity After Cyclic Shifts

class Solution(object):
    def areSimilar(self, mat, k):
        m = len(mat)
        n = len(mat[0])
        
        k = k % n   # Important optimization
        
        for i in range(m):
            for j in range(n):
                if i % 2 == 0:
                    # left shift
                    if mat[i][j] != mat[i][(j + k) % n]:
                        return False
                else:
                    # right shift
                    if mat[i][j] != mat[i][(j - k + n) % n]:
                        return False
        
        return True
```

```python
# 31. Next Permutation

class Solution(object):
    def nextPermutation(self, nums):
        n = len(nums)
        i = n - 2
        while i >= 0 and nums[i] >= nums[i + 1]:
            i -= 1

        if i >= 0:
            j = n - 1
            while nums[j] <= nums[i]:
                j -= 1

            nums[i], nums[j] = nums[j], nums[i]

        left, right = i + 1, n - 1
        while left < right:
            nums[left], nums[right] = nums[right], nums[left]
            left += 1
            right -= 1
        
```
