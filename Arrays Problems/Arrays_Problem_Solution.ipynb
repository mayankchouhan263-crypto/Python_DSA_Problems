```python
#2. Find the Maximum & Minimum Element
#GFG link "https://www.geeksforgeeks.org/problems/find-minimum-and-maximum-element-in-an-array4428/1"
# Test cases passed 1111/1111

class Solution1:
    def getMinMax(self, arr):
        Min = min(arr)
        Max = max(arr)
        return [Min,Max]

#OR 
class Solution2:
    def getMinMax(self, arr):
        Min = arr[0]
        Max = arr[0]
        for i in arr:
            if i< Min:
                Min = i
            if i> Max:
                Max = i
        return [Min,Max]

```

```python
#3. Find the Sum of Elements

class Solution:
    def arraySum(self, arr):
        total = 0
        for num in arr:
            total += num
        return total


```

```python
#4. Find the Second Largest Element
# GFG link https://www.geeksforgeeks.org/problems/second-largest3735/1
#Test Cases Passed 1120 / 1120
class Solution:
    def getSecondLargest(self, arr):
        max_val = -1
        second_max = -1
        
        for num in arr:
            if num > max_val:
                second_max = max_val
                max_val = num
            elif num > second_max and num != max_val:
                second_max = num
        
        return second_max

```

```python
#5. Count Frequency of Elements
#GFG Link "http://geeksforgeeks.org/dsa/counting-frequencies-of-array-elements/"
#Test Cases Passed 1112 / 1112

class Solution:
    def countFreq(self, arr):
        freq = {}

        for num in arr:
            if num in freq:
                freq[num] += 1
            else:
                freq[num] = 1

        result = [[key, value] for key, value in freq.items()]
        return result

```

```python
#6. Check if Array is Sorted
#GFG Link "https://www.geeksforgeeks.org/problems/check-if-an-array-is-sorted0701/1"
#Test Cases = 1112/1112

class Solution:
    def isSorted(self, arr) -> bool:
        result = True
        for i in range(0,len(arr)-1):
            if arr[i]>arr[i+1]:
                result = False
        return result
            
```
