```python
#2. Find the Maximum & Minimum Element
#GFG link "https://www.geeksforgeeks.org/problems/find-minimum-and-maximum-element-in-an-array4428/1"
# Test cases passed 1111/1111

class Solution1:
    def getMinMax(self, arr):
        Min = min(arr)
        Max = max(arr)
        return [Min,Max]

#OR 
class Solution2:
    def getMinMax(self, arr):
        Min = arr[0]
        Max = arr[0]
        for i in arr:
            if i< Min:
                Min = i
            if i> Max:
                Max = i
        return [Min,Max]

```

```python
#4. Find the Second Largest Element
# GFG link https://www.geeksforgeeks.org/problems/second-largest3735/1
#Test Cases Passed 1120 / 1120
class Solution:
    def getSecondLargest(self, arr):
        max_val = -1
        second_max = -1
        
        for num in arr:
            if num > max_val:
                second_max = max_val
                max_val = num
            elif num > second_max and num != max_val:
                second_max = num
        
        return second_max

```

```python
#5. Count Frequency of Elements
#GFG Link "http://geeksforgeeks.org/dsa/counting-frequencies-of-array-elements/"
#Test Cases Passed 1112 / 1112

class Solution:
    def countFreq(self, arr):
        freq = {}

        for num in arr:
            if num in freq:
                freq[num] += 1
            else:
                freq[num] = 1

        result = [[key, value] for key, value in freq.items()]
        return result

```

```python
#6. Check if Array is Sorted
#GFG Link "https://www.geeksforgeeks.org/problems/check-if-an-array-is-sorted0701/1"
#Test Cases = 1112/1112

class Solution:
    def isSorted(self, arr) -> bool:
        result = True
        for i in range(0,len(arr)-1):
            if arr[i]>arr[i+1]:
                result = False
        return result
            
```

```python
#7. Rotate Array by k Positions: Rotate the array to the right by k positions.
#gfg link = https://www.geeksforgeeks.org/problems/rotate-array-clockwise/1
#test cases = 1115 / 1115
class Solution(object):
    def rotate(self, nums, k):
        k = k % len(nums)   # handle k > len(nums)
        nums[:] = nums[-k:] + nums[:-k]

          
```

```python
#8. Find Pair with Given Sum: Find a pair of elements that adds up to a target sum.
#leetcode linkk = "https://leetcode.com/problems/two-sum/submissions/1923981711/"
#63 / 63 testcases passed
class Solution:
    def twoSum(self, nums, target):
        arr = [(num, i) for i, num in enumerate(nums)]
        arr.sort(key=lambda x: x[0])
        left = 0
        right = len(arr) - 1
        while left < right:
            current_sum = arr[left][0] + arr[right][0]

            if current_sum == target:
                return [arr[left][1], arr[right][1]]
            elif current_sum < target:
                left += 1
            else:
                right -= 1
        return []
```

```python
#9. Remove Duplicates from Array: Remove duplicates from the array while maintaining order.
#Test Cases Passed   1115 / 1115
#gfg link = "https://www.geeksforgeeks.org/problems/remove-duplicate-elements-from-sorted-array/1"
class Solution:
    def removeDuplicates(self, arr):
        n = len(arr)
        
        if n == 0:
            return []
        
        i = 0  
        
        for j in range(1, n):
            if arr[j] != arr[i]:
                i += 1
                arr[i] = arr[j]
        
        return arr[:i+1]
```

```python
#10. Merge Two Sorted Arrays
#59 / 59 testcases passed
#leecode = "https://leetcode.com/problems/merge-sorted-array/submissions/1924142420/"
class Solution(object):
    def merge(self, nums1, m, nums2, n):
        i = m - 1
        j = n - 1
        k = m + n - 1

        while(i >= 0 and j >= 0):
            if(nums1[i] > nums2[j]):
                nums1[k] = nums1[i]
                i= i-1
            else:
                nums1[k] = nums2[j]
                j-=1
            k-=1
        
        while(j >= 0):
            nums1[k] = nums2[j]
            j-=1
            k-=1
```
