```python
#4. Find the Second Largest Element
# GFG link https://www.geeksforgeeks.org/problems/second-largest3735/1
#Test Cases Passed 1120 / 1120
class Solution:
    def getSecondLargest(self, arr):
        max_val = -1
        second_max = -1
        
        for num in arr:
            if num > max_val:
                second_max = max_val
                max_val = num
            elif num > second_max and num != max_val:
                second_max = num
        
        return second_max

```

```python
#5. Count Frequency of Elements
#GFG Link "http://geeksforgeeks.org/dsa/counting-frequencies-of-array-elements/"
#Test Cases Passed 1112 / 1112

class Solution:
    def countFreq(self, arr):
        freq = {}

        for num in arr:
            if num in freq:
                freq[num] += 1
            else:
                freq[num] = 1

        result = [[key, value] for key, value in freq.items()]
        return result

```

```python
#8. Find Pair with Given Sum: Find a pair of elements that adds up to a target sum.
#leetcode linkk = "https://leetcode.com/problems/two-sum/submissions/1923981711/"
#63 / 63 testcases passed
class Solution:
    def twoSum(self, nums, target):
        arr = [(num, i) for i, num in enumerate(nums)]
        arr.sort(key=lambda x: x[0])
        left = 0
        right = len(arr) - 1
        while left < right:
            current_sum = arr[left][0] + arr[right][0]

            if current_sum == target:
                return [arr[left][1], arr[right][1]]
            elif current_sum < target:
                left += 1
            else:
                right -= 1
        return []
```

```python
#9. Remove Duplicates from Array: Remove duplicates from the array while maintaining order.
#Test Cases Passed   1115 / 1115
#gfg link = "https://www.geeksforgeeks.org/problems/remove-duplicate-elements-from-sorted-array/1"
class Solution:
    def removeDuplicates(self, arr):
        n = len(arr)
        
        if n == 0:
            return []
        
        i = 0  
        
        for j in range(1, n):
            if arr[j] != arr[i]:
                i += 1
                arr[i] = arr[j]
        
        return arr[:i+1]
```

```python
#10. Merge Two Sorted Arrays
#59 / 59 testcases passed
#leecode = "https://leetcode.com/problems/merge-sorted-array/submissions/1924142420/"
class Solution(object):
    def merge(self, nums1, m, nums2, n):
        i = m - 1
        j = n - 1
        k = m + n - 1

        while(i >= 0 and j >= 0):
            if(nums1[i] > nums2[j]):
                nums1[k] = nums1[i]
                i= i-1
            else:
                nums1[k] = nums2[j]
                j-=1
            k-=1
        
        while(j >= 0):
            nums1[k] = nums2[j]
            j-=1
            k-=1
```

```python
#13. Find Duplicates in an Array
#leetcode link = https://leetcode.com/problems/find-all-duplicates-in-an-array/submissions/1924429652/
#29 / 29 testcases passed
class Solution(object):
    def findDuplicates(self, nums):
        nums.sort()
        dup  = []
        for i in range(1,len(nums)):
            if nums[i-1] == nums[i] and nums[i] not in dup:
                dup.append(nums[i])
        return dup
```

```python
# 17. Find the Leader Elements: An element is a leader if it is greater than all elements to its right.
#Test Cases Passed   1111 / 1111
#gfg link "https://www.geeksforgeeks.org/problems/leaders-in-an-array-1587115620/1"
class Solution:
    def leaders(self, arr):
        leaders = []
        max = 0
        for i in range(len(arr) - 1, -1, -1):
            if max<=arr[i]:
                max = arr[i]
                leaders.append(max)
        leaders.reverse()
            
        return leaders  
```

```python
#19. Find Subarray with Given Sum.
#leetcode link = https://leetcode.com/problems/subarray-sum-equals-k/submissions/1924767831/
#93 / 93 testcases passed
class Solution(object):
    def subarraySum(self, nums, k):
        count = 0
        prefix_sum = 0
        freq = {0: 1}
        
        for num in nums:
            prefix_sum += num
            
            if prefix_sum - k in freq:
                count += freq[prefix_sum - k]
            
            freq[prefix_sum] = freq.get(prefix_sum, 0) + 1
        
        return count
```

```python
#23. Maximum Sum Subarray (Kadane's Algorithm)
#leetcode = https://leetcode.com/problems/maximum-subarray/
# 210 / 210 testcases passed
class Solution(object):
    def maxSubArray(self, nums):
        Max = nums[0]
        prefix_sum = 0
        
        for i in nums:
            prefix_sum+=i
            Max=max(prefix_sum,Max)
            if(prefix_sum<0):
                prefix_sum=0
        return Max
```

```python
#24. Rearrange Array Alternately: Rearrange an array such that elements alternate between the largest and smallest.
#https://www.geeksforgeeks.org/problems/-rearrange-array-alternately-1587115620/1
#1110 / 1110 testcases passes
class Solution:
    def rearrange(self, arr):
        arr.sort()
        n = len(arr)
        
        max_idx = n - 1
        min_idx = 0
        
        max_elem = arr[-1] + 1  
        
        for i in range(n):
            if i % 2 == 0:
                arr[i] += (arr[max_idx] % max_elem) * max_elem
                max_idx -= 1
            else:
                arr[i] += (arr[min_idx] % max_elem) * max_elem
                min_idx += 1
        
        
        for i in range(n):
            arr[i] //= max_elem
```

```python
#26. Find Peak Element: A peak element is greater than its neighbors. Find one such element.
# gfg link https://www.geeksforgeeks.org/problems/peak-element/1
#Test Cases Passed   1115 / 1115
class Solution:   
    def peakElement(self, nums):
        n = len(nums)

        if n == 1:
            return 0
            
        if nums[0] > nums[1]:
            return 0

        for i in range(1, n-1):
            if nums[i] > nums[i-1] and nums[i] > nums[i+1]:
                return i

        return n-1
```

```python
#27. Find the First Missing Positive: Find the smallest positive integer missing in the array.
# gfg link https://www.geeksforgeeks.org/problems/smallest-positive-missing-number-1587115621/1
#Test Cases Passed  1125 / 1125

class Solution(object):
    def firstMissingPositive(self, nums):
        n = len(nums)
        i = 0
        
        while i < n:
            correct_idx = nums[i] - 1
            if 1 <= nums[i] <= n and nums[i] != nums[correct_idx]:
                nums[i], nums[correct_idx] = nums[correct_idx], nums[i]
            else:
                i += 1
        
        for i in range(n):
            if nums[i] != i + 1:
                return i + 1
        
        return n + 1
```

```python
#29. Find the Longest Consecutive Sequence: Find the length of the longest consecutive sequence of integers.
#gfg link "https://www.geeksforgeeks.org/problems/longest-consecutive-subsequence2449/1"
#Test Cases Passed 1111 / 1111
class Solution(object):
    def longestConsecutive(self, nums):
        if len(nums)==0:
            return 0
        s = set(nums)
        longest = 0
        for i in s:
            if i-1 not in s:
                current = i
                length = 1
                while current+1 in s:
                    current += 1
                    length += 1
                longest = max(longest, length)
        return longest
```

```python
#30. Product of Array Except Self
# leetcode link = "https://leetcode.com/problems/product-of-array-except-self/"
#  24 / 24 testcases passed
class Solution(object):
    def productExceptSelf(self, nums):
        prefix_arr = [1] * len(nums)
        suffix_arr = [1] * len(nums)
        for i in range(1,len(nums)):
            prefix_arr[i] = prefix_arr[i-1] * nums[i-1]
        for i in range(len(nums)-2,-1,-1):
            suffix_arr[i] = suffix_arr[i+1]*nums[i+1]
        for i in range(len(nums)):
            prefix_arr[i] =prefix_arr[i]*suffix_arr[i]
        return prefix_arr
```
