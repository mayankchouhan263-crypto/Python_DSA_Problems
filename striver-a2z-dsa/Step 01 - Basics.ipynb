```python
#ALL-DIVISORS
#GFG Question link = "https://www.geeksforgeeks.org/problems/sum-of-all-divisors-from-1-to-n4738/1"
#1120 / 1120 Testcases passed
class Solution:
    def sumOfDivisors(self, n):
    	ans = 0
        for i in range(1, n+1):
            k = n // i
            ans += (i * k)
        return ans
```

```python
# ARMSTRONG-NUMBER

class Solution:
    def armstrongNumber(self, n):
        num_digits = len(str(n))   
        temp = n
        total = 0
        
        while temp > 0:
            digit = temp % 10
            total += digit ** num_digits
            temp //= 10  
        
        return "true" if total == n else "false"
```

```python
# BASIC-RECURSION

# SUM OF FIRST N NUMBERS
class Solution:
    def findSum(self, n):
        if n==0:
            return 0
        return n + self.findSum(n-1)
    
# FACTORIAL OF N
class Solution:
    
    def factorial(self, n: int) -> int:

        if n==0:
            return 1
        return n * self.factorial(n-1)

# FIBONACCI SERIES -> 1, 2, 3, 5, 8, 13, .......
class Solution(object):
    def fib(self, n):
        if n == 0:
            return 0
        elif n <= 2:
            return 1
        return self.fib(n-1) + self.fib(n-2)

    
```

```python
#CHECK-PALINDROME
#https://www.geeksforgeeks.org/problems/palindrome0746/1
#	Test Cases  1111 / 1111
class Solution:
    def isPalindrome(self, x):
        if x<0:
            return False
        s= str(x)

        if s == s[::-1]:
            return True
        else:
            return False

```
